```python
import pandas as pd

# -----------------------------
# Sample Dataset
# -----------------------------
data = {
    "First_Name": ["Hussain", "Ali", "Ahmed", "Sara", "Ayesha"],
    "Last_Name": ["Kashif", "Raza", "Khan", "Iqbal", "Malik"],
    "Age": [22, 30, 27, 24, 29],
    "Salary": [50000, 80000, 65000, 72000, 90000],
    "Department": ["IT", "HR", "IT", "Finance", "HR"],
    "Join_Date": [
        "2022-01-15",
        "2021-05-20",
        "2020-08-10",
        "2023-03-12",
        "2019-11-01"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# -----------------------------
# FEATURE ENGINEERING
# -----------------------------

# 1. Full Name Feature
df["Full_Name"] = df["First_Name"] + "_" + df["Last_Name"]

# 2. Age Group Feature
df["Age_Group"] = df["Age"].apply(
    lambda x: "Young" if x < 25 else "Adult"
)

# 3. Bonus Feature (10% of salary)
df["Bonus"] = df["Salary"] * 0.10

# 4. Total Salary After Bonus
df["Total_Salary"] = df["Salary"] + df["Bonus"]

# 5. High Salary Flag
df["High_Salary"] = df["Salary"].apply(
    lambda x: 1 if x > 70000 else 0
)

# 6. Convert Join Date to datetime
df["Join_Date"] = pd.to_datetime(df["Join_Date"])

# 7. Extract Year
df["Join_Year"] = df["Join_Date"].dt.year

# 8. Extract Month
df["Join_Month"] = df["Join_Date"].dt.month

# 9. Experience Years
current_year = 2026
df["Experience_Years"] = current_year - df["Join_Year"]

# 10. Salary Category
df["Salary_Category"] = pd.cut(
    df["Salary"],
    bins=[0, 60000, 80000, 100000],
    labels=["Low", "Medium", "High"]
)

# -----------------------------
# DISPLAY FINAL DATA
# -----------------------------
print("\n===== FINAL FEATURE ENGINEERING DATASET =====\n")
print(df)
```
